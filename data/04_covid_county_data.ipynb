```python
df2 = pd.read_json("counties_pop.json", orient="values")
```

```python
from collections import defaultdict
```

```python
codes = defaultdict(dict)
for d in df2["features"]:
    countyID = d["properties"]["GEO_ID"]
    code = d["properties"]["STATE"]
    countyName = d["properties"]["NAME"]
    state = code_to_state[code]
    codes[state][countyName] = countyID
```

```python
def get_county_cases():
    dict_cases = {}
    used_states = []
    for _,(state, county, cases, _) in df.iterrows():
        if not state in used_states:
            for county, geo in codes[state].items():
                dict_cases[geo] = (state, county, 0)
            used_states.append(state)
        try:
            geo_id = codes[state][county]
            dict_cases[geo_id] = (state, county, cases)
        except KeyError:
            print(state, county, cases)
            print("County not found")
            raise
    return dict_cases
```

```python
dict_cases = get_county_cases()
```

```python
def print_county_cases(dict_cases):
    used_states = []
    print("let dataCovidCounty = {")
    for geo_id, (state, county, cases) in dict_cases.items():
        if state not in used_states:
            print(f"\n\n//{state.upper()}")
            used_states.append(state)
        print(f"//{county}, {state}")
        print(f"'{geo_id}': {cases},")
    print("}")
```

```python
print_county_cases(dict_cases)
```
